```python
import dlib
import cv2
import numpy as np
import mysql.connector
```

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")  # Tải từ dlib
```

```python
def extract_facial_features(image_path):
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    faces = detector(gray)
    if len(faces) == 0:
        return None

    for face in faces:
        landmarks = predictor(gray, face)

        # Lấy tọa độ landmark
        points = np.array([(landmarks.part(i).x, landmarks.part(i).y) for i in range(68)])

        # 1. Khoảng cách hai mắt
        eye_distance = np.linalg.norm(points[42] - points[39])

        # 2. Chiều dài cung mày
        left_eyebrow_length = np.linalg.norm(points[17] - points[21])
        right_eyebrow_length = np.linalg.norm(points[22] - points[26])

        # 3. Khoảng cách giữa hai lông mày
        brow_distance = np.linalg.norm(points[21] - points[22])

        # 4. Chiều rộng miệng
        mouth_width = np.linalg.norm(points[48] - points[54])

        # 5. Chiều cao miệng
        mouth_height = np.linalg.norm(points[51] - points[57])

        # 6. Độ mở mắt (mắt trái và mắt phải)
        left_eye_opening = np.linalg.norm(points[37] - points[41])
        right_eye_opening = np.linalg.norm(points[43] - points[47])

        # 7. Chiều dài sống mũi
        nose_length = np.linalg.norm(points[27] - points[33])

        # 8. Chiều rộng mũi
        nose_width = np.linalg.norm(points[31] - points[35])

        # 9. Tỉ lệ chiều rộng và chiều cao khuôn mặt
        face_width = np.linalg.norm(points[0] - points[16])
        face_height = np.linalg.norm(points[8] - points[27])
        face_ratio = face_width / face_height if face_height != 0 else 0

        # 10. Độ dốc hàm
        jaw_slope = np.arctan2(points[8][1] - points[0][1], points[8][0] - points[0][0])

        return {
            "path": image_path,
            "eye_distance": eye_distance,
            "left_eyebrow_length": left_eyebrow_length,
            "right_eyebrow_length": right_eyebrow_length,
            "brow_distance": brow_distance,
            "mouth_width": mouth_width,
            "mouth_height": mouth_height,
            "left_eye_opening": left_eye_opening,
            "right_eye_opening": right_eye_opening,
            "nose_length": nose_length,
            "nose_width": nose_width,
            "face_ratio": face_ratio,
            "jaw_slope": jaw_slope
        }
```

```python
def save_face_features(data):
    # Kết nối đến MySQL
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="root",
        database="face_img_db"
    )
    cursor = conn.cursor()

    # Câu lệnh INSERT
    query = """
    INSERT INTO face_features (
        path, eye_distance, left_eyebrow_length, right_eyebrow_length,
        brow_distance, mouth_width, mouth_height, left_eye_opening,
        right_eye_opening, nose_length, nose_width, face_ratio, jaw_slope
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """

    values = (
        data["path"], data["eye_distance"], data["left_eyebrow_length"], 
        data["right_eyebrow_length"], data["brow_distance"], data["mouth_width"],
        data["mouth_height"], data["left_eye_opening"], data["right_eye_opening"],
        data["nose_length"], data["nose_width"], data["face_ratio"], data["jaw_slope"]
    )

    # Thực hiện câu lệnh
    cursor.execute(query, values)
    conn.commit()

    print("Data inserted successfully!")

    # Đóng kết nối
    cursor.close()
    conn.close()
```

```python
for i in range(1,331):
    img_path = f"image/male ({i}).jpg"
    img = extract_facial_features(img_path)
    if img:
        save_face_features(img)
    
```
